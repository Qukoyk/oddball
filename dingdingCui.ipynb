```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from google.colab import drive
import csv

drive.mount('/content/drive')

list1 = []
list2 = []
listRate = []
listE = []

file = "/content/drive/My Drive/dingdingCui/data.csv"

list1 = np.loadtxt(file, delimiter=',', skiprows=1, usecols=6)
list2 = np.loadtxt(file, delimiter=',', skiprows=1, usecols=7)

with open(file) as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        listE.append(int(row['E'].replace('E','')))
        pass
    pass

rows = 0

for i in list1:
    rows = rows + 1
    pass

print(str(rows)+' rows detected'+'\n')

for i in range(rows):
    listRate.append(list2[i]/list1[i])
    pass

plt.scatter(list1,listRate)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel("Action 1")
ax.set_ylabel("Convert Rates")
ax.set_zlabel("E number")
plt.tight_layout()

ax.scatter(list1,listRate,listE)

plt.show()
```
